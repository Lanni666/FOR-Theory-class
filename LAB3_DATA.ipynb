```python
#LAB3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
# grab the data from online, Excel is fine
url = 'https://www.qogdata.pol.gu.se/data/qog_bas_cs_jan24.xlsx'
df = pd.read_excel(url)
```

```python
df.head()
```

```python
print("Q1")
print('Countries with higher HDI are expected to have higher life expectancy due to better healthcare, education, and living standards.')
print("So, i choose Dependent Variable: Life Expectancy at Birth (wdi_lifexp)")
print("This represents the average life expectancy in years for a country's population.")
print("And Independent Variable: Human Development Index (HDI) (undp_hdi)")
```

```python
# Drop rows with missing values for the selected variables
regression_data = df[['wdi_lifexp', 'undp_hdi']].dropna()

# Run the bivariate regression
lifeexp_hdi_model = smf.ols(formula='wdi_lifexp ~ undp_hdi', data=regression_data).fit()

# Print the regression summary
print(lifeexp_hdi_model.summary())
```

```python

print("\n   - R-squared is 0.834")
print("     * This indicates that 83.4% of the variance in life expectancy is explained by HDI.")
print("\n   - Coefficient for undp_hdi: 44.93")
print("     * this means for each 0.1 increase in HDI, life expectancy increases by approximately 4.49 years.")
print("     * This confirms a strong positive relationship between HDI and life expectancy.")


print("\n   - Intercept is 39.43")
print("     * This imply life expectancy when HDI equals 0 is 39.43 years.")
print("     * While this intercept is not meaningful in real-world terms")

print("   - The p-value for the HDI coefficient is <0.001.")
print("     * This means the relationship between HDI and life expectancy is highly statistically significant.")

print("   - Adjusted R-squared: 0.834")
print("     * This value confirms that the model fits the data very well.")
print("   - F-statistic: 927.2 with a p-value of 9.07e-74.")
print("     * Indicates that the overall model is highly statistically significant.")

print("\n  **Conclusion**")
print("   - **Yes, the results fit expectations.**")
print("     * A strong positive association between HDI and life expectancy.")
print("     * Because countries with higher HDI means have better access to healthcare, education, and living standards, contributing to longer life expectancy.")
print("     * The high R-squared value and statistical significance support this conclusion.")

print("   - and the regression strongly supports the hypothesis that HDI is a strong predictor of life expectancy.")
print("   - Policies aimed at improving HDI, such as investments in healthcare, education, and income, are likely to result in increased life expectancy.")
```

```python
print('Q2: CHoose Healthcare Expenditure as Percentage of GDP (wdi_expedu)')
#This variable reflects a country's investment in healthcare relative to its GDP.
#may have realationship with 2 variable
# Drop rows with missing values for the relevant variables
regression_data = df[['wdi_lifexp', 'undp_hdi', 'wdi_expedu']].dropna()

# Run the regression with an additional variable
lifeexp_hdi_health_model = smf.ols(formula='wdi_lifexp ~ undp_hdi + wdi_expedu', data=regression_data).fit()

# Print the regression summary
print(lifeexp_hdi_health_model.summary())
```

```python
print("   - R-squared is 0.849")
print("     * This indicates that 84.9% of the variance in life expectancy is explained by HDI and healthcare expenditure combined.")

print("   - Coefficient for HDI (undp_hdi): 45.78")
print("     * which means 0.1 increase in HDI is associated with a 4.578-year increase in life expectancy.")
print("     * This coefficient is positive and statistically significant (p < 0.001), suggesting HDI is a major driver of life expectancy.")

print("\n   - Coefficient for Healthcare Expenditure (wdi_expedu): -0.124")
print("     * A 1% increase in healthcare expenditure as a percentage of GDP is associated with a decrease of 0.124 years in life expectancy.")
print("     * This result is not statistically significant (p = 0.275), as the p-value is above the 0.05 threshold.")
print("   - Healthcare expenditure is not statistically significant, suggesting it does not have a clear linear effect on life expectancy in this model.")

print("In Conclusion")
print("It is only partly work")
print("     * Adding healthcare expenditure did not improve the explanatory power of the model much")
print("     * The coefficient for HDI remained largely unchanged, indicating healthcare expenditure does not mediate the effect of HDI on life expectancy.")
print("     * But the negative and insignificant coefficient for healthcare expenditure is surprising and may reflect complex, non-linear relationships.")
print("This shows policies aimed at improving HDI are likely to have the most substantial impact on life expectancy.")
```
