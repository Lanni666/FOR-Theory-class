```python
# LAB2
# 1-- Recode 2 different variables into new categories. They can both be continuous-ish or both be nominal-ish, or one of each. Tell me what you did and explain the variable(s).
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
url = 'https://www.qogdata.pol.gu.se/data/qog_bas_cs_jan24.xlsx'
df = pd.read_excel(url)
```

```python
df.head()
```

```python
# I choose the Continuous Variable: `wdi_gdpcappppcon2017` (GDP per capita in constant 2017 PPP dollars)
# and the Nominal Variable: `p_polity2` (Polity2 democracy score).

mean_value = df['wdi_gdpcappppcon2017'].mean()
std_dev = df['wdi_gdpcappppcon2017'].std()
median_value = df['wdi_gdpcappppcon2017'].median()
min_value = df['wdi_gdpcappppcon2017'].min()
max_value = df['wdi_gdpcappppcon2017'].max()
print("Choose variable wdi_gdpcappppcon2017")
print(f"Mean: {mean_value}")
print(f"Standard Deviation: {std_dev}")
print(f"Median: {median_value}")
print(f"Minimum: {min_value}")
print(f"Maximum: {max_value}")


mean_value2 = df['p_polity2'].mean()
std_dev2 = df['p_polity2'].std()
median_value2 = df['p_polity2'].median()
min_value2 = df['p_polity2'].min()
max_value2 = df['p_polity2'].max()
print("Choose variable p_polity2")
print(f"Mean: {mean_value2}")
print(f"Standard Deviation: {std_dev2}")
print(f"Median: {median_value2}")
print(f"Minimum: {min_value2}")
print(f"Maximum: {max_value2}")
```

```python
print("After check the distribution of 2 varible, I will recoded two variables from the dataset:")
print("\n1. Continuous Variable: 'wdi_gdpcappppcon2017' (GDP per capita in constant 2017 PPP dollars).")
print("   - I grouped this variable into three categories:")
print("     * 'Low GDP': GDP per capita less than 10,000.")
print("     * 'Medium GDP': GDP per capita between 10,000 and 30,000.")
print("     * 'High GDP': GDP per capita greater than 30,000.")
print("\n2. Nominal Variable: 'p_polity2' (Polity2 democracy score).")
print("   - I grouped this variable into three categories:")
print("     * 'Autocracy': Polity2 score less than or equal to 0.")
print("     * 'Partial Democracy': Polity2 score between 1 and 5.")
print("     * 'Full Democracy': Polity2 score between 6 and 10.")
```

```python
# Recode GDP per capita
df['gdp_category'] = pd.cut(df['wdi_gdpcappppcon2017'], bins=[-float('inf'), 10000, 30000, float('inf')],
                            labels=['Low GDP', 'Medium GDP', 'High GDP'])

# Recode Democracy (Polity2 score)
df['democracy_category'] = pd.cut(df['p_polity2'], bins=[-float('inf'), 0, 5, 10],
                                  labels=['Autocracy', 'Partial Democracy', 'Full Democracy'])

# Display the first few rows with new variables
df[['wdi_gdpcappppcon2017', 'gdp_category', 'p_polity2', 'democracy_category']].head()

print("\nHere is a preview of the recoded variables:")
print(df[['wdi_gdpcappppcon2017', 'gdp_category', 'p_polity2', 'democracy_category']].head())
#Some country have missing value
```

```python
#**2-- Use one (or both) of your recoded variables to do a cross-tabulation. Explain your results.**
# Perform cross-tabulation of GDP category and democracy category
crosstab = pd.crosstab(df['gdp_category'], df['democracy_category'], margins=True)

# Display the cross-tabulation table
print("Cross-tabulation of GDP Categories and Democracy Categories:")
print(crosstab)
```

```python
#explaination
print("My Results of the Cross-Tabulation of GDP Categories and Democracy Categories:\n")

print("1. This cross-tabulation shows the distribution of countries across GDP categories "
      "('Low GDP', 'Medium GDP', and 'High GDP') and Democracy categories "
      "('Autocracy', 'Partial Democracy', and 'Full Democracy').\n")

print("2. Key Observations:")
print("   - Low GDP Countries:")
print("     * Majority (31 out of 65) of Low GDP countries are classified as 'Full Democracy'.")
print("     * A significant number (21 out of 65) are 'Autocracy'.")
print("     * Only 13 Low GDP countries are 'Partial Democracy'.\n")

print("   - Medium GDP Countries:")
print("     * The largest group (34 out of 53) of Medium GDP countries are 'Full Democracy'.")
print("     * 12 are classified as 'Autocracy'.")
print("     * Only 7 are 'Partial Democracy'.\n")

print("   - High GDP Countries:")
print("     * Most High GDP countries (31 out of 39) are 'Full Democracy'.")
print("     * Only 7 are 'Autocracy', and just 1 is 'Partial Democracy'.\n")

print("3. Overall Totals:")
print("   - Total countries: 157")
print("   - 'Full Democracy' is the most frequent category, with 96 countries.")
print("   - 'Autocracy' accounts for 40 countries, while 'Partial Democracy' accounts for 21.\n")

print("**Conclusion:**")
print("There is a clear relationship between GDP and Democracy:")
print("   - Countries with higher GDP tend to be classified as 'Full Democracy'.")
print("   - Countries with lower GDP are more likely to be 'Autocracy'.")
print("   - 'Partial Democracy' is relatively uncommon across all GDP categories.")
```

```python
#3
# Drop rows with missing values for the variables
correlation_data = df[['wdi_gdpcappppcon2017', 'p_polity2']].dropna()

# Compute the Pearson correlation
correlation = correlation_data['wdi_gdpcappppcon2017'].corr(correlation_data['p_polity2'])

# Print the correlation result
print("**Correlation Analysis**")
print("I computed the Pearson correlation between GDP per capita (wdi_gdpcappppcon2017) and the Polity2 Democracy Score (p_polity2):\n")
print(f"Correlation Coefficient: {correlation:.2f}")
```

```python

print("1. The correlation coefficient of 0.14 indicates a weak positive relationship between GDP per capita and democracy levels.")
print("   - This means that, on average, as GDP per capita increases, the democracy score also tends to increase slightly.")
print("2. However, the relationship is weak, suggesting that other factors may have a stronger influence on democracy levels.")
print("3. This result could reflect the diversity of countries in the dataset, where high GDP does not always correspond to higher democracy scores.")
print("\n**Conclusion:**")
print("While there is a slight positive relationship between economic development (measured by GDP per capita) and democracy, "
      "the weak correlation suggests that GDP alone does not strongly predict a country's democracy score. ")
print('Other variables may play a significant role.')
```

```python
# 4 Find extreme cases for GDP per capita
max_gdp = df['wdi_gdpcappppcon2017'].max()
min_gdp = df['wdi_gdpcappppcon2017'].min()

# Get details of countries with max and min GDP
max_gdp_case = df[df['wdi_gdpcappppcon2017'] == max_gdp]
min_gdp_case = df[df['wdi_gdpcappppcon2017'] == min_gdp]

# Print the results
print("**Q4 Extreme Cases Analysis**\n")
print(f"1. Country with the Highest GDP per Capita:\n{max_gdp_case[['cname', 'wdi_gdpcappppcon2017']]}\n")
print(f"2. Country with the Lowest GDP per Capita:\n{min_gdp_case[['cname', 'wdi_gdpcappppcon2017']]}\n")

print("So:")
print("The 2 extreme value are high, the differences between Luxembourg and Burundi highlight the vast disparities in economic development across countries.")
print("This emphasizes the importance of tailored economic and policy interventions for low-GDP countries like Burundi.")
print("1. Luxembourg has the highest GDP per capita in the dataset:")
print("   - This is reasonable, this country have a strong economic development, small population, and focus on financial services, which contribute to its wealth.")
print("2. Burundi has the lowest GDP per capita in the dataset:")
print("   - This also reasonable, because this country may face significant economic challenges, including widespread poverty, underdeveloped infrastructure, and political instability.")

```
